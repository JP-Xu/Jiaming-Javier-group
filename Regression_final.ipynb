# Regression

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit as cf
```

## Cation: - pyridinium

```python
number_hydrocarbons = np.array([4, 3, 2, 1])
net_charge = np.array([0.00346608672462140, 
                       0.00363861282823367, 
                       0.00386863246403481,
                       0.00386863246403481
                      ])
```

```python
def complicated_function(x, a, b, c):
    return (a * x * x + b * x + c)
optimal_parameters, covariance = cf(complicated_function, number_hydrocarbons, net_charge)
```

```python
def report(optimal_parameters, covariance):
    parameter_errors = np.sqrt(np.diag(covariance))
    for i in range(len(optimal_parameters)):
        print("Parameter {}: {} +/- {} (1 st. dev.)".format(i,
                                                            optimal_parameters[i],
                                                            parameter_errors[i]))

    # Plot the data
    plt.plot(number_hydrocarbons, net_charge, 'o', label='data')

    # Make a new x array with 50 points for smoother lines
    x_many_points = np.linspace(number_hydrocarbons.min(), number_hydrocarbons.max(),50)
    
    plt.plot(x_many_points, complicated_function(x_many_points, *optimal_parameters), label='fitted')
    plt.legend(loc='best')
    plt.xlabel('Hydrocarbon group length')
    plt.ylabel('Net charge')
    plt.title('- Pyridinium')
    plt.show()

report(optimal_parameters, covariance)
```
